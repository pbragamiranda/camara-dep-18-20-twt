```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
dep_followers_fp = '/home/blackmamba/code/pbragamiranda/dep-followers/data/dep_followers.csv'
dep_followers = pd.read_csv(dep_followers_fp)

dep_info_fp = '/home/blackmamba/code/pbragamiranda/dep-followers/data/dep_description.csv'
dep_info = pd.read_csv(dep_info_fp)
```

```python
print(dep_followers.columns)
print("---")
print(dep_info.columns)
```

```python
followers = pd.pivot_table(dep_followers, index=['twitter_account'], 
                          columns=['date'], values='followers')

followers.reset_index(level=0, inplace=True)
followers.columns.date = 'id'
```

```python
followers.head()
```

```python
deputies = pd.merge(dep_info, followers, on='twitter_account')
```

```python
deputies.head()
```

```python
def diff(x, y):
    return x - y
```

```python
deputies['diff-28-27'] = diff(deputies['28/02/2021'], deputies['27/02/2021'])
deputies['diff-01-28'] = diff(deputies['01/03/2021'], deputies['28/02/2021'])
deputies['diff-02-01'] = diff(deputies['02/03/2021'], deputies['01/03/2021'])
```

```python
deputies.sort_values(by='diff-02-01', ascending=True).head(10)
```

```python
deputies[deputies['diff-02-03'].notnull()].sort_values(by='diff-02-03', ascending=True).tail(10)
```

```python
deputies.groupby(['party'])['28/02/2021'].sum().reset_index().sort_values(by='28/02/2021', ascending=False)
```

```python
deputies.groupby(['UF'])['28/02/2021'].sum().reset_index().sort_values(by='28/02/2021', ascending=False)
```
